```python
def format_salary(salary):
    return "{:,.0f}".format(salary)
```

```python
def convert_salary_to_number(salary_str):
    salary_str = salary_str.replace('Lacs P.A.', '').replace('Lacs', '').replace('P.A.', '').strip()
    return float(salary_str) * 100000
```

```python
def process_salary(salary):
    if 'Not Disclosed' in salary or 'Unpaid' in salary:
        return None, salary
    elif '-' in salary:
        # Handle salary ranges
        lower, upper = salary.split('-')
        lower = convert_salary_to_number(lower)
        upper = convert_salary_to_number(upper)
        return format_salary(math.floor(lower)), format_salary(math.ceil(upper))
    else:
        # Handle fixed salaries
        fixed = convert_salary_to_number(salary)
        return None, format_salary(fixed)
```

```python
df['Lower Salary'] = None
df['Salary Cap'] = None
```

```python
for index, row in df.iterrows():
    try:
        lower, upper = process_salary(row['Salary'])
        df.at[index, 'Lower Salary'] = lower
        df.at[index, 'Salary Cap'] = upper
    except ValueError:
        df.at[index, 'Lower Salary'] = None
        df.at[index, 'Salary Cap'] = row['Salary']
```
