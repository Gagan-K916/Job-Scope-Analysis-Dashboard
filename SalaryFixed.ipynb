```python
import re
import math

def format_salary(salary):
    return "{:,.0f}".format(salary)

def convert_salary_to_number(salary_str, multiply=False):
    salary_str = re.sub(r'[^\d.]+', '', salary_str)
    value = float(salary_str)
    return value * 100000 if multiply else value

def process_salary(salary):
    original_salary = salary
    salary = salary.replace('Lacs P.A.', '').replace('Lacs', '').replace('P.A.', '').strip()
    
    if 'Not Disclosed' in salary or 'Unpaid' in salary:
        return None, salary
    elif '-' in salary:
        # Handle salary ranges
        lower, upper = salary.split('-')
        multiply = 'Lacs' in original_salary or 'Cr' in original_salary
        lower = convert_salary_to_number(lower, multiply)
        upper = convert_salary_to_number(upper, multiply)
        return format_salary(math.floor(lower)), format_salary(math.ceil(upper))
    elif '/month' in salary.lower():
        # Handle per month salaries
        multiply = 'Lacs' in original_salary or 'Cr' in original_salary
        fixed = convert_salary_to_number(salary, multiply)
        return None, format_salary(fixed)
    else:
        # Handle fixed salaries
        multiply = 'Lacs' in original_salary or 'Cr' in original_salary
        fixed = convert_salary_to_number(salary, multiply)
        return None, format_salary(fixed)

# Create new columns
df['Lower Salary'] = None
df['Salary Cap'] = None

# Populate the new columns
for index, row in df.iterrows():
    try:
        lower, upper = process_salary(row['Salary'])
        df.at[index, 'Lower Salary'] = lower
        df.at[index, 'Salary Cap'] = upper
    except ValueError:
        df.at[index, 'Lower Salary'] = None
        df.at[index, 'Salary Cap'] = row['Salary']

```
