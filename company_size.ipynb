```python
import pandas as pd
```

```python
df = pd.read_csv("company_info.csv", encoding='utf-8', encoding_errors='replace')
```

```python
df.head()
```

```python
import numpy as np

# Function to clean and categorize company size
def categorize_company_size(size):
    # Check for NaN values
    if pd.isna(size):
        return np.nan 
    size = str(size)
    if 'Nov' in size or 'Oct' in size:
        return 'Small' 
    if '+' in size:
        lower = int(size.replace('+', ''))
        delta = lower + 1 
    else:
        try:
            lower, upper = map(int, size.split('-'))
            delta = upper - lower
        except ValueError:
            return np.nan  
    
    if delta <= 200:
        return 'Small'
    elif delta <= 5000:
        return 'Medium'
    else:
        return 'Large'

df['company size category'] = df['Company Size'].apply(categorize_company_size)
```

## MAKING NEW COLUMNS ##

```python
import pandas as pd
import numpy as np

df1 = pd.read_csv("company_info.csv", encoding='utf-8', encoding_errors='replace')
```

```python
def categorize_company_size(size):
    # Check for NaN values
    if pd.isna(size):
        return np.nan  
    
    size = str(size)
    
    # Handling special cases
    if 'Nov' in size or 'Oct' in size:
        return 'Small'  
    if '+' in size:
        lower = int(size.replace('+', ''))
        delta = lower + 1 
    else:
        try:
            lower, upper = map(int, size.split('-'))
            delta = upper - lower
        except ValueError:
            return np.nan  
    
    if delta <= 200:
        return 'Small'
    elif delta <= 5000:
        return 'Medium'
    else:
        return 'Large'

df1['company size category'] = df1['Company Size'].apply(categorize_company_size)
df1['Small'] = df1['company size category'] == 'Small'
df1['Medium'] = df1['company size category'] == 'Medium'
df1['Large'] = df1['company size category'] == 'Large'


df1.to_csv('company_info_newcolumns.csv', index=False)
```
